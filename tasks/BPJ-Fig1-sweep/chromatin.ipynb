```python
import json

import colorcet
import matplotlib.pyplot as plt
import numpy as np

from lib.plots import set_axis, plot_rect
```

```python
task_id = "7-acetylation_sweep"
config_id = 71

with open(f"{task_id}/_configs/config-{config_id}.json") as file:
    config = json.load(file)

meta = config["@meta"]
chain_length = meta["config_params"]["chain_length"]
info = meta["config_info"]
enhancer_intervals = info["islands"]
print(info["acetylation_level"])
```

```python
fig, ax = plt.subplots(figsize=(2.5, 0.3))

for start, end in enhancer_intervals:
    plot_rect(ax, start, 0, end, 0.5, facecolor="#e44", edgecolor="none", zorder=1)

ax.plot([0, chain_length], [0.5, 0.5], color="k", zorder=2)

s = "φ = {:g}".format(info["acetylation_level"])
ax.text(0, 0.55, s, fontsize="small", ha="left", va="bottom")

ax.set_xlim(0, chain_length)
ax.set_ylim(-0.3, 1.3)

set_axis(ax, "none")

fig.savefig(f"_figs/chromatin-{task_id}-{config_id}.svg")

pass
```
