```python
import glob
import json

import polars as pl
```

```python
task_id = "task-7"

metrics_table = pl.read_csv(f"{task_id}/metrics.csv")
```

```python
for config_id, separation_score in metrics_table.select("config_id", "separation_score").iter_rows():
    with open(f"{task_id}/_configs/config-{config_id}.json") as file:
        config = json.load(file)

    chain, = config["chains"]
    loops = chain["static_loops"]

    loop_ends = [(loop["pair"][0], loop["pair"][1]) for loop in loops]
    print(
        "{}: {} -- {:.4g}".format(
            config_id,
            " ".join(f"{i}:{j} ({abs(i - j)})" for i, j in loop_ends),
            separation_score,
        ),
    )
```
